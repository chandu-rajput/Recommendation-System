````python
def create_tags(x):
    x = x.copy()

    x['tags'] = (
        (x['title'] + " ") * 3 +
        x['overview'] + " " +
        (x['genres'] + " ") * 3 +
        (x['keywords'] + " ") * 2 +
        (x['cast'] + " ") * 2 +
        (x['director'] + " ") * 3 +
        x['tagline'] + " " +
        x['original_language'] + " " +
        x['year']
    )


    x['tags'] = (
        x['tags']
        .str.lower()
        .str.replace(r'\s+', ' ')
        .str.strip()
    )

    return x
````

````python
import re
import nltk
from nltk.stem import WordNetLemmatizer

nltk.download('wordnet')
````

````python
lemmatizer = WordNetLemmatizer()
````

````python
def preprocess_tags(text):

  # removing punctuation
  text = re.sub(r'[^a-zA-Z0-9\s]', '', text)

  # lemmatizer each word
  text = ' '.join([lemmatizer.lemmatize(word) for word in text.split()])

  return text
````

````python
movies_df['tags'] = movies_df['tags'].apply(preprocess_tags)
````

**Why TF-Iraw_movies_df over CountVectorizer:**

```
CountVectorizer → just counts word frequency
                  "action" appears 500x → gets too dominant

TF-Iraw_movies_df          → balances frequency vs uniqueness
                  common words like "action" get lower weight
                  unique words get higher weight → better similarity

````python
from sklearn.feature_extraction.text import TfidfVectorizer
````

````python
# Initialize TF-Iraw_movies_df Vectorizer to convert text tags into numerical vectors
# We use unigrams and bigrams to capture phrases like "Science Fiction"

tfidf = TfidfVectorizer(
    stop_words='english',
    max_features=10000,
    ngram_range=(1, 2),
    min_df=2,
)
````

````python
movies_df.reset_index(drop=True,inplace=True)
````

````python
tfidf_matrix = tfidf.fit_transform(movies_df['tags'])
````

````python
print(tfidf_matrix.shape)
````

````python
indices = pd.Series(
    movies_df.index,
    index=movies_df['title'].str.lower().str.strip()
).drop_duplicates()
````

## IMDB Formula:

**WR = (v/(v+m)) × R + (m/(v+m)) × C**
*  v = movie's vote count
*  m = minimum votes required (threshold)
*  R = movie's vote average
*  C = mean vote average across all movies

````python
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics.pairwise import cosine_similarity
````

````python
# Constants calculated from the entire dataset
C = movies_df['vote_average'].mean()
m = movies_df['vote_count'].quantile(0.60)
````

````python
def weighted_rating(x, C=C, m=m):
  v = x['vote_count']
  R = x['vote_average']
  return (v / (v + m)) * R + (m / (m + v)) * C
````

**Full flow summary:**
```
Input movie
    → find its vector
    → compare vs all 4800 movies (cosine similarity)
    → take top 100 similar
    → filter + deduplicate
    → score each by similarity + IMDB quality
    → normalize both scores to same scale
    → blend 60/40
    → return top 5

````python
def recommend(movie, top_n=5, pool_size=100):
    """
    Hybrid Recommendation System
    ----------------------------
    1. Finds content similarity using TF-Iraw_movies_df and Cosine Similarity.
    2. Filters a candidate pool of 100 movies.
    3. Calculates Weighted Ratings to prioritize quality movies.
    4. Blends scores (60% Similarity / 40% Rating) for the final result.
    """

    # Standardize input and find movie index
    movie_lower = movie.lower().strip()
    if movie_lower not in indices:
        return "Movie not found in database."

    idx = indices[movie_lower]

    # Step 1: Compute Cosine Similarity for the target movie
    sim_scores = cosine_similarity(tfidf_matrix[idx], tfidf_matrix)[0]

    # Step 2: Create a candidate pool (Top 100 similar movies)
    # We take 100 to ensure we have enough "good" movies to rank by rating
    sim_indices = sim_scores.argsort()[::-1][1:pool_size + 1]
    rec = movies_df.iloc[sim_indices].copy()
    rec['sim_score'] = sim_scores[sim_indices]

    # Step 3: Data Cleaning
    rec = rec[rec['title'].str.lower() != movie_lower]
    rec = rec.drop_duplicates(subset='title')

    # Step 4: Apply Weighted Rating
    rec['weighted_rating'] = rec.apply(weighted_rating, axis=1)

    # Step 5: Normalization
    # We scale both scores to [0, 1] so they can be added fairly
    scaler = MinMaxScaler()
    rec[['norm_sim', 'norm_rating']] = scaler.fit_transform(
        rec[['sim_score', 'weighted_rating']]
    )

    # Step 6: The Hybrid Blend
    # 0.6 and 0.4 are the weights for Content vs. Quality
    rec['hybrid_score'] = (0.6 * rec['norm_sim']) + (0.4 * rec['norm_rating'])

    # Step 7: Final Sort
    cols_to_show = ['title', 'year', 'genres', 'sim_score', 'hybrid_score']
    return rec.sort_values(by='hybrid_score', ascending=False)[cols_to_show].head(top_n)
````

````python
recommend('the terminator')
````
